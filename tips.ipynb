```python
import seaborn as sns

df = sns.load_dataset('tips')
```

Mostrar estatisticas descritivas

```python
df.info()
```

```python
df.describe().round(2)
```

Mostrar distribuição dos dados

```python
import matplotlib.pyplot as plt
sns.histplot(df['total_bill'], kde=True)
plt.title('Distribuição do total conta')
plt.show()
```

Mostrar outliers

```python
sns.boxplot(df['total_bill'])
plt.title('Distribuição do total conta')
plt.show()
```

Analisar correlação das váriaveis numéricas

```python
df.corr(numeric_only=True)
```

Mostrar relação entre as váriaveis

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='Blues')
plt.title('Mapa de correlação')
plt.show()
```

Mostrar relaçaão total X tip

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Relação entre total e gorjeta')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='size', data=df)
plt.title('Relação entre total e gorjeta')
plt.show()
```

```python
sns.scatterplot(x='tip', y='size', data=df)
plt.title('Relação entre size e gorjeta')
plt.show()
```

Analisar agrupamentos

```python
df.groupby('sex')['tip'].max()
```

```python
sns.barplot(x='sex', y='tip',data=df)
plt.title('Média de gorjeta por sexo')
plt.show()
```

```python
df.groupby('sex')['total_bill'].max()
sns.barplot(x='sex', y='total_bill',data=df)
plt.title('Maior conta por sexo')
plt.show()
```

Contagem de categorias

```python
sns.countplot(x='time', data=df)
plt.title('Quantidade de clientes por período')
plt.show()
```

Analisar multiplos valores

```python
sns.scatterplot(x='total_bill', y='tip', hue='size', size='size', data=df)
plt.title('Conta x gorjeta por tamanho de mesa')
plt.show()
```
