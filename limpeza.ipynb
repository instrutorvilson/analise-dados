```python
import pandas as pd

df = pd.read_csv('dados.csv')

print(df.to_string())
```

```python
new_df = df.dropna()
print(new_df.to_string())
```

```python
new_df['Date'] = pd.to_datetime(new_df['Date'], format='mixed')

print(new_df.to_string())
```

```python
new_df.loc[7, 'Duration'] = 45
print(new_df.to_string())
```

```python
new_df.drop_duplicates(inplace = True)
print(new_df.to_string())
```
